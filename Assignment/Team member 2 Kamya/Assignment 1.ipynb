## 7. Concatenate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
#Code:
a = np.array([1, 2, 3])

b = np.array([4, 5, 6])
g = np.concatenate((a, b), axis = None)
print(g)
```
